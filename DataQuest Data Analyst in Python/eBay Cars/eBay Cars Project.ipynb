### 'registration_year' observations

The max of 9999 for registration year is bad data and would need to be examined and possibly removed.  The same is true of the minimum year of 1000.  To give it the most inclusive filter, we could only allow registration dates bewtween 1885 (year car was invented) and the 2016 (year the data is from).

```python
registration_adjusted_autos = cleaned_data[cleaned_data["registration_year"].between(1885,2016)]
cleaned_data = registration_adjusted_autos
print(registration_adjusted_autos['registration_year'].describe())
registration_adjusted_autos['registration_year'].value_counts(normalize=True, dropna=False).sort_index()
```

'registration_year' now has a much more expected spread with dates from the early 1900s not have many cars registered while more cars are registered more recently.

### Aggregate by Car Brand

```python
cleaned_data['brand'].value_counts(normalize=True, dropna=False)
```

I'm going to aggregate on any brands with more than 1% of the listings.

```python
brand_list = cleaned_data['brand'].value_counts(normalize=True, dropna=False).sort_index()
partial_brand_list = brand_list[brand_list >= .01]
print(partial_brand_list)
mean_price_dict = {}
mean_mileage_dict = {}

for brand in partial_brand_list.index:
    rows = cleaned_data.loc[cleaned_data['brand'] == brand]
    mean_price_dict[brand] = rows['price'].mean()
    mean_mileage_dict[brand] = rows['odometer_km'].mean()
```

Prices of the cars with more than 1% of the listings range from about $2,500 (renault) to $9,380(audi).

```python
price_series = pd.Series(mean_price_dict)
km_series = pd.Series(mean_mileage_dict)

price_mileage_df = pd.DataFrame(price_series, columns=['mean_price'])
price_mileage_df['mean_km'] = km_series

price_mileage_df.sort_values(['mean_price'], ascending=False)
```

There is not a correlation between the mean price and the mean odometer reading.
